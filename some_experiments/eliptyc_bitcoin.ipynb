```python
import numpy as np
import matplotlib.pyplot as plt

def main():
#   graph of SECP256k1 y**2 = x**3 + 7
#   y**2 = x**3 + a*x + b  
    a = 0
    b = 7

    y, x = np.ogrid[-10:10:100j, -10:10:100j]
    plt.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a - b, [0])
    plt.grid()
    plt.show()

if __name__ == '__main__':
    main()
```
